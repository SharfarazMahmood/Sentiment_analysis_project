```
import pandas as pd
import numpy as np
import os
import fnmatch
```

```
data_dir = "/content/drive/My Drive/Work_FTFL_GIGATECH/Sentiment_Analysis_Group_1"
file_list = [ f for f in os.listdir(data_dir) if fnmatch.fnmatch(f , 'Group-1-*')]
print(file_list)

## read and merge anotations from all files
dataframe = pd.read_excel(os.path.join(data_dir, "Group-1- Sharfaraz.xlsx"), sheet_name="Sheet1"  )
dataframe = dataframe['Text']
for file in file_list:
    df = pd.read_excel(os.path.join(data_dir, file), sheet_name="Sheet1"  )
    file= file.replace("Group-1- ", "").replace(".xlsx","")
    # print(file)
    dataframe = pd.concat([dataframe, df['Polarity ']], axis=1)
    dataframe = dataframe.rename(columns={"Polarity ": file}) 
dataframe = dataframe.replace(np.nan, 'Objective', regex=True)

## task1: change "Strongly" & "Negatively" catagories to single catagory
dataframe = dataframe.replace("Strongly Positive", 'Positive', regex=True)
dataframe = dataframe.replace("Weakly Positive", 'Positive', regex=True)
dataframe = dataframe.replace("Strongly Negative", 'Negative', regex=True)
dataframe = dataframe.replace("Weakly Negative", 'Negative', regex=True)


## task2: find annotation using majority voting
annotation_vote_count = dataframe.iloc[:, 1:].apply(pd.Series.value_counts, axis=1).fillna(0)
annotation_list = list(annotation_vote_count.columns)
final_annotation = []
evenly_voted = []
# print(annotation_list)
for index in annotation_vote_count.index:
    vote_count = list(annotation_vote_count.loc[index])
    if vote_count.count(max(vote_count)) == 1 :
        final_annotation.append( annotation_list[vote_count.index(max(vote_count))]  )
    else:
        final_annotation.append( "evenly_voted" )
        evenly_voted.append(index)
    # print(vote_count, max(vote_count), vote_count.count(max(vote_count)) , final_annotation[index])

dataframe["Polarity "] = final_annotation
output_file = "annotation_results_group_1_(Sharfaraz_Harun_Ruhi_Shabiba_Lutfunnesa)_Zubayear.xlsx"
dataframe.to_excel( os.path.join(data_dir, output_file), index=False)
# dataframe.to_excel("/content/drive/My Drive/Work_FTFL_GIGATECH/annotation_rexults.xlsx", index=False) 

dataframe.iloc[:25,]
```
